```python
import pandas as pd
import numpy as np
```

```python
loan = pd.read_csv('LoanStats3a.csv', skiprows=1)
loan.head()
```

```python
loan.shape
```

```python
loan.info()
```

```python
half_count = len(loan)//2
loan = loan.dropna(thresh=half_count, axis=1)
loan = loan.drop(['desc'], axis=1)
```

```python
loan.shape
```

```python
loan.drop_duplicates()  # 删除重复项
```

```python
loan.shape
```

```python
loan.columns
```

```python
loan = loan.drop(["funded_amnt", "funded_amnt_inv", "grade", "sub_grade", "emp_title", "issue_d",
                              "total_rec_int", "total_rec_late_fee", "recoveries", "collection_recovery_fee", 
                              "last_pymnt_d", "last_pymnt_amnt"], axis=1)
```

```python
loan = loan.drop(["zip_code", "out_prncp", "out_prncp_inv", "total_pymnt", "total_pymnt_inv", "total_rec_prncp"], axis=1)
loan.shape
```

```python
orig_columns = loan.columns
drop_columns = []
for col in orig_columns:
    col_series = loan[col].dropna().unique()
    if len(col_series) == 1:
        drop_columns.append(col)
loan = loan.drop(drop_columns, axis=1)
print(drop_columns)
```

```python
null_counts = loan.isnull().sum()  # 统计每个特征项的空值个数
print(null_counts)
```

```python
loan = loan.drop(['pub_rec_bankruptcies'],axis=1)  # 认为这个项缺失太多，还是删除
loan = loan.dropna(axis=0)  # 这里对所有有缺失值的样本都删除
null_counts_2 = loan.isnull().sum()
print(null_counts_2)
```

```python
status_replace = {
    'loan_status': {
        'Fully Paid': 1,
        'Charged Off': 0,
        'Does not meet the credit policy. Status:Fully Paid': 1,
        'Does not meet the credit policy. Status:Charged Off': 0
    }
}
loan = loan.replace(status_replace)
loan.loan_status.value_counts()
```

```python
print(loan.dtypes.value_counts())
```

```python
loan.info()
```

```python
object_colunms_df = loan.select_dtypes(include=['object'])
print(object_colunms_df.iloc[0])
```

```python
cols = ['home_ownership', 'verification_status', 'emp_length', 'debt_settlement_flag', 'term', 'addr_state','purpose', 'title']
for c in cols:
    print(loan[c].value_counts())
```

```python
loan = loan.drop(["last_credit_pull_d", "earliest_cr_line", "addr_state", "title"], axis=1)
mapping_dict = {
    "emp_length": {
        "10+ years": 10,
        "9 years": 9,
        "8 years": 8,
        "7 years": 7,
        "6 years": 6,
        "5 years": 5,
        "4 years": 4,
        "3 years": 3,
        "2 years": 2,
        "1 year": 1,
        "< 1 year": 0,
    }
}
loan['int_rate'] = loan['int_rate'].str.rstrip('%').astype('float')
loan['revol_util'] = loan['revol_util'].str.rstrip('%').astype('float')
loan = loan.replace(mapping_dict)
```

```python
cat_columns = ["home_ownership", "verification_status", "purpose", "term", 'debt_settlement_flag']
dummy_df = pd.get_dummies(loan[cat_columns]) # one-hot encoding
loan = pd.concat([loan, dummy_df], axis=1)
loan = loan.drop(cat_columns, axis=1)
```
